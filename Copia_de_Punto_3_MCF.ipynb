```python
#Punto 3
import numpy as np
print("Punto 3")
print("Problema de diseño")

# Parámetros
Q= 0.07
L=600
dh=20
epvc=0.0000015
ehierro=0.00015
k=10
vc_agua=0.00000114

print("Asumiendo D=0.1860")

D=0.1860
Re = 4*Q/(vc_agua*np.pi*D)
f = 1./( (-1.8*np.log10((6.9/Re) + np.power((epvc/D)/3.7,1.11) ))**2 )

li = dh
ld = (f*(600/D)+10)*(0.0392/(96.82081917*(D**4)))
dl = ld - li


#Ciclo para calcular D
print("Para tubería de PVC")
i=1
for i in range(1,7):
  D = D - 0.0006
  print(" ")
  print("Iteracion # : ",i)
  print("Diámetro de prueba : ",D)
  Re = 4*Q/(vc_agua*np.pi*D)
  f = 1./( (-1.8*np.log10((6.9/Re) + np.power((epvc/D)/3.7,1.11) ))**2 )
  ld = (f*(600/D)+10)*(0.0392/(96.82081917*(D**4)))
  dl = ld - li
  print("ld - li = ",dl)
  print(" ")
  i = i + 1

print("Para tubería de PVC")
print("Finalmente, el lado izquierdo es: ",li)
print("Y el lado derecho es: ",ld)
print("Obtenido para un diámetro D = ",D)
print(" ")

print("Para tubería en Hierro")
D_hierro=0.1860
j=1
for j in range(1,5):
  D_hierro = D_hierro + 0.0019
  print(" ")
  print("Iteracion # : ",j)
  print("Diámetro de prueba : ",D_hierro)
  Re = 4*Q/(vc_agua*np.pi*D_hierro)
  f = 1./( (-1.8*np.log10((6.9/Re) + np.power((ehierro/D_hierro)/3.7,1.11) ))**2 )
  ld = (f*(600/D_hierro)+10)*(0.0392/(96.82081917*(D_hierro**4)))
  dl = ld - li
  print("ld - li = ",dl)
  print(" ")
  j = j + 1

print("Para tubería en Hierro")
print("Finalmente, el lado izquierdo es: ",li)
print("Y el lado derecho es: ",ld)
print("Obtenido para un diámetro D = ",D_hierro)
print(" ")


```
