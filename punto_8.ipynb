```python
print("Punto 8")
import numpy as np

#Parámetros
dh = 18
l = 87
d = 0.0508
k = 11
e = 0.0000015
A = np.pi*((d**2)/4)
vc_a = 0.00000114

print("Asumimos f = 0.04")
f = 0.04
#Despejamos Q en función de f
#Q = np.sqrt((-471682.2672)/((-0.000000000000200448*f)-(128747520000)))

#Calculamos Q, Re y f hasta convergencia

for i in range(3):
  Q = np.sqrt((-471682.2672)/((-0.000000000000200448*f)-(128747520000)))
  Re = (Q*d)/(vc_a*A)
  f = 1./( (-1.8*np.log10((6.9/Re) + np.power((e/d)/3.7,1.11) ))**2 )
  Q = np.sqrt((-471682.2672)/((-0.000000000000200448*f)-(128747520000)))
  print('Iteración : ',i+1)
  print('f = ',f)
  print('Q = ',Q)
  print('Re = ',Re)
  print(" ")

#Ahora calculamos las HB1 y HB2

HB1 = 20-(0.00005*(Q**2))
HB2 = 20-(0.00005*(Q**2))

print("R/ Entonces HB1 y HB2 tendrán el mismo valor de: HB1=",HB1,"m y HB2=",HB2,"m")
print("Mientras que Q valdrá: Q=",Q,"m³/s")
```
